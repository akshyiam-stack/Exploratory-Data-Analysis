```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ft = pd.read_csv(r"C:\Users\Prince\Downloads\global_sports_footwear_sales_2018_2026.csv")
ft.head(10)
```

```python
ft.columns = ft.columns.str.title()
ft.head(10)
```

```python
ft.info()
```

```python
ft['Order_Date'] = pd.to_datetime(ft['Order_Date'])
ft['Year'] = ft['Order_Date'].dt.year
```

```python
ft.info()
```

```python
ft.isnull().sum()
```

```python
ft.duplicated().sum()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(ft['Revenue_Usd'] , kde = True)
plt.title('sales Distribution of Revenue (USD)')
plt.xlabel('Revenue (USD)') 
plt.ylabel('Frequency')
plt.show()
```

```python
rev = ft.groupby('Year')['Revenue_Usd'].sum().reset_index().sort_values('Revenue_Usd' , ascending = False)

sns.barplot(x = 'Year' ,  y = 'Revenue_Usd' , data = rev , palette = 'rocket')
plt.title('Revenue (USD) By year')
plt.xlabel('Year')
plt.ylabel('Revenue (USD)')
plt.show()

```

```python
sns.lineplot(
    x = 'Year' , 
    y = 'Revenue_Usd' , 
    data = ft , 
)
plt.title('Revenue (USD) By year')
plt.xlabel('Year')
plt.ylabel('Revenue (USD)') 
plt.show()
```

which country had the heighest revenue

```python
country_revenue = ft.groupby('Country')['Revenue_Usd'].sum().reset_index().sort_values(by = 'Revenue_Usd', ascending = False)
sns.barplot(x = 'Country' ,
             y = 'Revenue_Usd' ,
               data = country_revenue ,
                 palette = 'mako')
plt.title('Revenue (USD) by country')
plt.xlabel('country')
plt.ylabel('Revenue (usd)')
plt.show() 
```

which brand had the heighest revenue

```python
brand_revenue = ft.groupby('Brand')['Revenue_Usd'].sum().reset_index().sort_values(by = 'Revenue_Usd', ascending=False)

sns.barplot(x = 'Brand' , 
            y = 'Revenue_Usd' ,
              data = brand_revenue , 
              palette = 'viridis')
plt.title('Revenue (Usd) by brand')
plt.xlabel('Brand')
plt.ylabel('Revenue (usd)')
plt.xticks(rotation =  45)
plt.show
```

which product category sells  the most

```python
product_sales = ft.groupby('Category') ['Revenue_Usd'].sum().reset_index().sort_values(by ='Revenue_Usd' , ascending = False)

sns.barplot(x = 'Category' ,
             y = 'Revenue_Usd' ,
               data = product_sales ,
                 palette = 'plasma')

plt.title('Revenue (Usd) by Product_category')
plt.xlabel('prodcut catory')
plt.ylabel('revenue (usd)')
plt.xticks(rotation = 45)
plt.show()
```

relationship b/w price and sales

```python
ft['Price'] = ft['Revenue_Usd']/ft['Units_Sold'] # dreatinng a new column price by dividing revenue by units sold
```

```python
sns.regplot(x = 'Price' , y = 'Revenue_Usd' , data = ft , color = 'Green')
plt.title('Revenue (Usd) vs price')
plt.xlabel('Price')
plt.ylabel('Revenue (Usd)')
plt.show()
```

detect outliers in Revenue_usd

```python
sns.boxplot(x = ft['Revenue_Usd'])
plt.title('outliers in  Revenue (USD)')
plt.show()
```

sales distribution by product category

```python
sale_prdct = ft.groupby('Category')['Units_Sold'].sum().reset_index().sort_values(by = 'Units_Sold' , ascending = False)

sns.histplot(sale_prdct['Units_Sold'] , kde = True)
plt.title('sales Distribution of Revenue (USD)')
plt.xlabel('Revenue (USD)') 
plt.ylabel('Frequency')
plt.show()
```

count of orders per product category

```python
order = ft['Category'].value_counts().sort_values(  ascending = False).index

sns.countplot(x = 'Category', data = ft , order=order , palette = 'rocket')
plt.title("count orders per category")
plt.xlabel('category')
plt.ylabel('count')
plt.show()
```

month wise sales

```python
ft['Month'] = ft['Order_Date'].dt.month 
```

```python
month_sales = ft.groupby('Month')['Revenue_Usd'].sum().reset_index()\
    .sort_values(by = 'Revenue_Usd' , ascending = False)

sns.lineplot(x = 'Month' , y = 'Revenue_Usd' , data = ft , color = 'red')
plt.title('Revenue (Usd) by month')
plt.xlabel('Month')
plt.ylabel('Revenue (USD)')
plt.xticks(rotation = 45)
plt.show()
```

sales by gender

```python
gender_sales = ft.groupby('Gender')['Revenue_Usd'].sum().reset_index().sort_values(by = 'Revenue_Usd' , ascending = False)

sns.barplot( x = 'Gender' , 
            y = 'Revenue_Usd' , 
            data = gender_sales , 
            palette = 'rocket') 
plt.title('Revenue (usd) by gender')
plt.xlabel('Gender')
plt.ylabel('Revnue (Usd)')
plt.show()
```

compare sales acroos years by different country

```python
sale_year = ft.groupby(['Year' , 'Country'])['Revenue_Usd'].sum().reset_index()
plt.figure(figsize=(12,6))

sns.barplot( x = 'Year', 
            y = 'Revenue_Usd' , 
            hue = 'Country' , 
            data = sale_year ,
            palette = 'dark:salmon_r')
plt.title('sales across years by country')
plt.xlabel('Year')
plt.ylabel('Revenue (usd)')
plt.xticks(rotation = 45)
plt.show()
```

sales across years by different category

```python
year_sales = ft.groupby([ 'Year' , 'Category'])['Revenue_Usd'].sum().reset_index().sort_values(by = 'Revenue_Usd' , ascending = False)
plt.figure(figsize=(12,6))

sns.barplot( x = 'Year' , 
            y = 'Revenue_Usd' , 
            hue = 'Category' , 
            data = year_sales , 
            palette = 'Blues')
plt.title('sales across years by category')
plt.xlabel('Year')
plt.ylabel('Revenue (usd)')
plt.xticks(rotation = 45)
plt.show()
```

monthly sales comparison by category

```python
monthly_category = ft.groupby(['Month' , 'Category'])['Revenue_Usd'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.barplot(x = 'Month' , 
            y = 'Revenue_Usd' ,  
            hue = 'Category' , 
            data = monthly_category , 
            palette = 'YlOrBr')

plt.title('sales across months by category')
plt.xlabel('Months')
plt.ylabel('Revenue (usd)')
plt.xticks(rotation = 45)
plt.show()
```

which Product Category Sells the Most Units

```python
highest_cateegory = ft.groupby('Category')['Units_Sold'].sum().reset_index().sort_values(by = 'Units_Sold' , ascending = False)
plt.figure(figsize=(12,6))

sns.barplot(x = 'Category' , 
            y = 'Units_Sold' , 
            data = highest_cateegory , 
            palette = 'coolwarm')

plt.title('category sells the most')
plt.xlabel('Category')
plt.ylabel('Units sold')
plt.xticks(rotation = 45)
plt.show()
```

Year-over-Year Growth in Revenue

```python
year_grender = ft.groupby('Year')['Revenue_Usd'].sum().reset_index()
plt.figure(figsize=(12,6))

sns.lineplot(x = 'Year' , 
             y = 'Revenue_Usd' , 
             data = year_grender ,
             palette = 'magma')
plt.title('Year-over-Grwoth')
plt.ylabel('Revenue (Usd)')
plt.xlabel('Year')
plt.xticks(rotation = 45)
plt.show()
```

Country Contributes the Highest Share of Total Revenue
